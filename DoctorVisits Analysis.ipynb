```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# read the dataset

```python
df=pd.read_csv("DoctorVisits - DA.csv")
df.head(25)
```

# Displaying the information about the columns of the dataset

```python
df.info()
```

# finding the number of people based on their illness

```python
df["illness"].value_counts(numeric_only=True)
```

# finding the number of people based on their age

```python
df["age"].value_counts()
```

# Visualizing ana analysing income

```python
y=list(df.income)
plt.boxplot(y)
plt.show()
```

# finding the no.of days of reduced activity of male and female seperately due to illness

```python
df.groupby(['gender','reduced']).mean()
```

```python
df.groupby(['age','income']).mean()
```

# visualize is there any missing values in the dataset based on a heat map

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

# finding out the different corelation between different variables

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Greens')
```

# Analyzing using scatter plot

```python
plt.figure(figsize=(8,8))
plt.scatter(x='age',y='illness',data=df)
plt.xlabel('age')
plt.ylabel('illness')
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x='income',y='gender',data=df)
plt.xlabel('income')
plt.ylabel('gender')
```

# count and visualize the number of males and females affected by illness

```python
sns.histplot(df.gender,bins=5)
```

# plot a horizontal bar chat to analyze the reduced days of activity due to illness based on gender

```python
db=df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightgreen'])
```

# Analyzung about the insurance

```python
# % of people getting govt insurance due to low income
label=['yes','no']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("% of people getting govt insurance due to low income")
plt.show()
# % of people having private health insurance
Y = df[df['private']=='yes']
N = df[df['private']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("% of people having private health insurance")
plt.show()
# % of people getting govt insurance due to old age,disability or veteran status
label=['yes','no']
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.show()

```
